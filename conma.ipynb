```python
import json
import tqdm
import pandas as pd
```

```python
from sklearn.metrics import confusion_matrix
```

```python
labels = ["Bonds", "Forward", "Funds", "Future", "MMIs", "Option", "Stocks", "Swap", "Equity Index", "Credit Index", "Securities restrictions", "Parametric schedules", "Debt pricing and yields", "Credit Events", "Stock Corporation", "Central Securities Depository", "Regulatory Agency"]
```

```python
base_path = '/home/chaofeng/Documents/vscode/finsim/data/kg/1/'
file = 'b1.json'
```

```python
data = pd.read_json(base_path+file)
```

```python
data
```

```python
y_true = data['label'].to_list()
```

```python
y_pred = []
for i in data['predicted_labels'].to_list():
    y_pred.append(i[0])
```

```python
con = confusion_matrix(y_true, y_pred, labels=labels)
```

```python

import seaborn as sn
```

```python
sn.heatmap(con, annot=True,xticklabels= labels, yticklabels=labels, cmap='YlGnBu', cbar=False)
```
